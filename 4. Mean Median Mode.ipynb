# Mean Median & Mode

## Mean Vs Median

Let's create some fake income data, 
centered around 27,000 with a normal distribution and standard deviation of 15,000, with 10,000 data points. 

compute the mean (average) - it should be close to 27,000:


```python
import numpy as np

incomes = np.random.normal(27000, 15000, 10000)
incomes
```

```python
incomes.mean()
```

```python
np.mean(incomes)
```

We can segment the income data into 50 buckets, and plot it as a histogram:

```python
import matplotlib.pyplot as plt
plt.hist(incomes, bins=50)
plt.show()
```

Now compute the median - since we have a nice, even distribution it too should be close to 27,000:

```python
np.median(incomes)
```

Now we'll add outliers income inequality!

```python
incomes = np.append(incomes, [1000000000])
```

The median won't change much, but the mean does:

```python
np.median(incomes) # little bit changes 
```

```python
np.mean(incomes) # previuosly it was 27331
```
