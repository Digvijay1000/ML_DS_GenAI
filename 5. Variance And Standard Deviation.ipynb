# Variance And Standard Deviation


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 50.0, 10000) # normal(loc(Mean)=0.0, scale(STD)=1.0, size=None)

plt.hist(incomes, 50)
plt.show()
```

```python
incomes.std()
```

```python
incomes.var()
```
